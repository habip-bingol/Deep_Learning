## Preprocessing of Data
- Train | Test Split, Scalling

```python
from sklearn.model_selection import train_test_split
```

```python
# X = df[['feature1', 'feature2']].values  # convert to array
# y = df['price'].values
```

```python
X = df[['feature1', 'feature2']] 
y = df['price']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# help(MinMaxScaler)
```

```python
scaler = MinMaxScaler()
```

```python
# Gradient Descent te 0 ile 1 arasnda ekmek iimizi kolaylatrr.
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Modelling

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dense, Activation
```

```python
# help(Sequential)
```

```python
# help(Dense)
```

```python
X_train.shape
```

```python
# First way to create model

model = Sequential([Input(shape = X_train.shape[1]),
                    Dense(units = 4, activation = "relu"), # first hidden layer
                    Dense(units = 4, activation = "relu"), # second hidden layer
                    Dense(units = 4, activation = "relu"), # third hidden layer
                    Dense(units = 1)])# last layer katman(single neuron because it is  regression problem. default activation is linear)

# Sequential: Ard arda bal olacak model tipimiz
# Input iine de yazabiliriz ayr ayr da yazabiliriz alttaki gibi
# Dense : Fully connected layer eklemek demek
# units=4 : 4 nron olacak(o layer da)
# activation : aktivasyon fonksiyonumuz "relu" olacak 
# input_dim = X_train.shape[1]  /  Input(shape = X_train.shape[1] : Giri datamn shape i (Bamsz feature(deiken) says kadar)
```

```python
# Second way to create model (recommended)

model = Sequential()

# model.add(Input(shape = X_train.shape[1]))
model.add(Dense(units = 4, activation = "relu", input_dim = X_train.shape[1])) # first hidden layer
model.add(Dense(units = 4, activation = "relu")) # second hidden layer
model.add(Dense(units = 4, activation = "relu")) # third hidden layer
model.add(Dense(units = 1)) # last layer katman(single neuron because it is  regression problem. default activation is linear)

# Bu saylar belirlemenin net bir yolu yok bunu dataya bakarak karar vereceiz
# Az feature var o yzden ok hidden layer a gerek yok. Aslnda 3 hidden layer da fazla
# Data bykse nronlar arttrlabilir, layerlar arttrlabilir vs
```

```python
model.summary()
# birinci satr
    # Birinci katmanda 4 nron varm
    # Batch_size olmad iin None veriyor
    # Param = 12  .. 4 eik ve 2(input layer) x 4(birinci hidden layer)=8 toplam 12
```

```python
model.weights
# Bias leri genelde ilk bata 0 alyor
# Weightleri random atyor.
```

### Choosing an optimizer and loss

Keep in mind what kind of problem you are trying to solve:

    # For a multi-class classification problem
    model.compile(optimizer='rmsprop',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    # For a binary classification problem
    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    # For a mean squared error regression problem
    model.compile(optimizer='rmsprop',
                  loss='mse')

```python
model.compile(optimizer = "rmsprop", loss = "mse")
```

```python
model.fit(x = X_train, y = y_train, batch_size = 32, epochs = 250)

# Her bir batch de 32 tane satr alacak
# 700(train datas satr says)/32 = 21*32 + 28 satr(22. batch)  --> Toplam 22 iterasyon 
# Her epoch da 22 iterasyon yapacak sonra
# 22 iterasyonun ortalamasn alyor sonu veriyor ve update ediyor

# NEML NOT: Hatann tekrar dzelmesi iin tekrar altrabiliriz ama "fit" den itibaren altrrsak
# .. nceden kard sonular zerinden gider. Bu salkl deil(local minimuma yakalanm olabiliriz ve oradan kamayz).
# .. Bunun iin "Sequential" ksmndan itibaren modeli yeniden (belki tekrar tekrar) altrmalyz. 

# Batch_size lar 2 nin ss olacak ekilde  semekte fayda var nk daha hzl alyor(Baka rakamlarda verebiliriz ama ok byk datalarda sknt olabilir)
```

```python
model.weights
# Eitim bittikten sonra initial a gre;
# Biasler deiti
# Weightler deiti
```

```python
model.history.history
# Her bir epoch daki loss lar gsteriyor
```

```python
pd.DataFrame(model.history.history)
```

## Model Performance

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
model.evaluate(X_train, y_train, verbose=0) # En son elde ettii MSE deerini gsteriyor(Train iin)
```

```python
model.evaluate(X_test, y_test, verbose=0) # En son elde ettii MSE deerini gsteriyor(Test iin)
```

```python
my_dict = {"Actual" : y_test, "Pred" : y_pred.reshape(-1,)}
compare = pd.DataFrame(my_dict)
compare.sample(10)
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:", mse, "\nrmse:", rmse)
```

```python
eval_metric(y_test, y_pred)
```

### Saving a Model

You can save a TensorFlow/Keras model using [`model.save()`](https://www.tensorflow.org/tutorials/keras/save_and_load#save_the_entire_model).

Keras saves models by inspecting their architectures. This technique saves everything:

- The weight values
- The model's architecture
- The model's training configuration (what you pass to the .compile() method)
- The optimizer and its state, if any (this enables you to restart training where you left off)

There are two ways to save a model in TensorFlow:
1. The [SavedModel format](https://www.tensorflow.org/tutorials/keras/save_and_load#savedmodel_format) (default).
2. The [HDF5 format](https://www.tensorflow.org/tutorials/keras/save_and_load#hdf5_format).



```python
final_model.save('model_fake_reg.h5')  # creates a HDF5 file 'my_model.h5'

# Peki ne kaydediyor? Alttakileri
"""
The weight values
The model's architecture
The model's training configuration (what you pass to the .compile() method)
The optimizer and its state, if any (this enables you to restart training where you left off)
"""
```
